```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the two images (you can replace these paths with your own images)
image1_url = '/content/ex5_clahe_1.png'
image2_url = '/content/ex5_do_nawab-100x100.png'

# Load images and check if they were loaded successfully
image1 = cv2.imread(image1_url, cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread(image2_url, cv2.IMREAD_GRAYSCALE)

if image1 is None or image2 is None:
    print("Error: Unable to load one or both images.")
else:
    # Perform normalization
    normalized_image1 = image1.astype(float) / 255.0  # Normalize image1 to the range [0, 1]
    normalized_image2 = image2.astype(float) / 255.0  # Normalize image2 to the range [0, 1]

    # Display the results
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(normalized_image1, cmap='gray'), plt.title('Normalized Image 1')
    plt.subplot(122), plt.imshow(normalized_image2, cmap='gray'), plt.title('Normalized Image 2')
    plt.show()
```

```python

# Load the two images
image1_url = '/content/ex5_clahe_1.png'
image2_url = '/content/ex5_do_nawab-100x100.png'

image1 = cv2.imread(image1_url, cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread(image2_url, cv2.IMREAD_GRAYSCALE)

# Perform histogram equalization
equalized_image1 = cv2.equalizeHist(image1)
equalized_image2 = cv2.equalizeHist(image2)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(equalized_image1, cmap='gray'), plt.title('Equalized Image 1')
plt.subplot(122), plt.imshow(equalized_image2, cmap='gray'), plt.title('Equalized Image 2')
plt.show()
```
